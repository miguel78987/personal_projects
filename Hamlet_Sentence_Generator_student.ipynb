# First step: tidy the text of the scene, breaking each sentence into an ordered list of the words in that sentence.

````python
def modifyText (aString):
    #Convert all letters to lowercase
    aString = aString.lower()

    #find nth instance of substring
    def find_nth(mainString, subString, n):
        start = mainString.find(subString)
        while start >= 0 and n > 1:
            start = mainString.find(subString, start+len(subString))
            n -= 1
        return start

    sentenceList = []

    startingSentenceIndex = 0
    #endingSentenceIndex will be found using find_nth function

    periodCount = 0
    questionCount = 0
    exclamationCount = 0

    #Split string by punctuation marks
    for c in aString:
        if c in '.?!':
            if c == '.':
                #Update nth instance of '.'
                periodCount += 1

                endingSentenceIndex = find_nth(aString, c, periodCount)

            if c == '?':
                #Update nth instance of '?'
                questionCount += 1

                endingSentenceIndex = find_nth(aString, c, questionCount)

            if c == '!':
                #Update nth instance of '!'
                exclamationCount += 1

                endingSentenceIndex = find_nth(aString, c, exclamationCount)
            
            #Slice the sentence
            sentenceString = aString[startingSentenceIndex:endingSentenceIndex]

            #Remove trailing whitespace from sentence
            sentenceString = sentenceString.strip()

            #Remove newline characters
            sentenceString = sentenceString.replace('\n', '')

            #Remove apostrophes
            if "'" in sentenceString:
                sentenceString = sentenceString.replace("'", '')

            #Remove quotation marks
            if '"' in sentenceString:
                sentenceString = sentenceString.replace('"', '')

            #Remove commas 
            if ',' in sentenceString:
                sentenceString = sentenceString.replace(',', '')


            #Split sentence into list of words
            wordList = sentenceString.split(' ')

            #Add the list of words in the sentence to sentence list
            sentenceList.append(wordList)


            #Update beggining of next sentence
            startingSentenceIndex = endingSentenceIndex + 1


    return sentenceList
````

# Second step: create a dictionary where the keys are words from the text, and the values are lists of the words that immediately follow the key words in the dialogue.

Now create a sequential pairs dictionary, called `hamdict`, with `key` being each word in `hamsplits` and `value` being the list of words that immediately following it. If no words ever follow the given word, then it should not appear as a key in `hamdict`.

````python
hamdict = {}

for sentence in hamsplits:

    #No follow up word in this sentence
    if len(sentence) == 1:
        continue

    for wordIndex in range(len(sentence)):
        word = sentence[wordIndex]

        #Check if there's a followup word
        #Check if wordIndex is the last index in the list
        if wordIndex != (len(sentence) - 1):
            followUpWord = sentence[wordIndex + 1]

            #Create dictionary key with list value
            if word not in hamdict.keys():
                hamdict[word] = []

            #Add followup word to dictionary
            #and avoid duplicate values
            if followUpWord not in hamdict[word]:
                hamdict[word].append(followUpWord)

#hamdict length should be 552
#print(len(hamdict)) 
````

# Third step: create a phrase generator that will use that dictionary by starting with a key word, randomly selecting a word from the list of words that follows it, adding that word to our phrase, and repeating the process.

Below, write a function, `hambot`, that takes a starter word from your dictionary and builds a Shakespearean-sounding phrase, returning it as a string. 

Your function should randomly select one of the words that follows your starter word, add it to your phrase, and then repeat the process with this new word. This process should continue until either your function reaches a word that is not a key in your dictionary, or the phrase reaches the specified length. Your function should then return this phrase as a string (all lowercase, with no punctuation).

The function takes two optional parameters, `length` and `starter`. `length` is an integer which is the number of words long your output phrase should be. If the `length` parameter is not supplied, your function should default to attempting to produce a 5-word phrase.

`starter` is the first word to start your phrase.  If `starter` is not supplied, your function should choose a word at random from among the keys in your dictionary.

As an example, if your function starts with the word "some," the second word in your phrase would be either "confession" or "danger," the two words that follow "some." If it selected "confession," the third word would be "of," which is the only word that follows "confession." The fourth word would be randomly chosen from the 38 words that follow "of." In other words,

```python
hambot(4, "some")
```

.... should return a string like, `"some confession of circumstance"`.

> Note: A useful Python function for choosing a random element of a list is `random.choice()`.

````python
import random

def hambot(length = 5, starter = False):
    #Assign random starter word if starter word is not supplied
    if starter == False:
        starter = random.choice(list(hamdict.keys()))
    
    nextWord = str(random.choice(list(hamdict[starter])))

    hamOutput = [starter] #This list will be converted to a string

    while length > 1:
        length -= 1

        #Add follow up word to list
        hamOutput.append(nextWord)

        if nextWord in hamdict.keys():
            nextWord = str(random.choice(list(hamdict[nextWord])))
        
        else:
            #Exit while loop if next word is not a dictionary key
            break
    
    hamOutput = ' '.join(hamOutput)

    return hamOutput
````
